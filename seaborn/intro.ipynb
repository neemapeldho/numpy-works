```python
"""Seaborn"""

# python data visualization library which write on top of matplotlib

# seaborn is more styled mean while matplotlib look outdated
```

```python
import seaborn
import pandas
from matplotlib import pyplot
```

lineplot

```python
days = ['sun','mon','tue','wed','thurs','fri','sat']

sales = [120,150,160,200,220,170,140]

data = pandas.DataFrame({ "Days" : days , "Sales" : sales})

seaborn.lineplot(data= data, x = "Days" , y = "Sales" , linestyle = "--", marker= "*")

pyplot.show()
```

```python
days = ['sun','mon','tue','wed','thurs','fri','sat']

sales = [120,150,160,200,220,170,140]

data = pandas.DataFrame({ "Days" : days , "Sales" : sales})

seaborn.lineplot(data= data, marker = "*")

pyplot.xlabel("days")

pyplot.ylabel("sales")

pyplot.title("sales and days")

pyplot.show()
```

boxplot

```python
employee_data = {
    "dept" : ["sales","IT","marketing","testing","Hr"],
    "salaries" : [20000,40000,15000,30000,35000]
}

df = pandas.DataFrame(employee_data)
df

seaborn.boxplot(x="dept",y="salaries",data=df,linecolor= "yellow")
pyplot.title("dept and salaries")
pyplot.show()
```

barplot

```python
subjects = ["physics","chemistry","biology","maths","history"]

marks = [70,80,90,96,35]

data = pandas.DataFrame({ "subjects" : subjects , "marks" : marks})

seaborn.barplot(data=data , x = "subjects" , y= "marks" , palette= "coolwarm" ,hue = "marks")

pyplot.title("student grade card")

pyplot.show()
```

histogram

```python
# histogram 
# seaborn.histplot

ages = [22,32,45,67,45,89,34,24,89,65,31,90,43,83,89]

seaborn.histplot(data=ages)

pyplot.title("number of people in specific age")

pyplot.xlabel("ages")

pyplot.show()
```
