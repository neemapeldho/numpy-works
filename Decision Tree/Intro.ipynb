Binary Classification
======================

If we have only two calsses to predict then it is called Binary Classification


In Binary Classification the maximum impurity is when both classes have equal proportion

To calculate the impurity in huge dataset we use to find Entropy

And the process is called measure of impurity or measure of randomness
