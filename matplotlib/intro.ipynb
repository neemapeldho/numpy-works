```python
from matplotlib import pyplot
```

```python
x = [2,3,4,5,6]
y = [8,9,10,11,12]

pyplot.plot(x,y)
pyplot.show()
```

```python
#plot , scattered , histogram  , bar , piechart
```

```python
# scattered >>> graph not using connecting lines just points

height = [20,30,40,50,60]
weight = [39,29,10,11,12]

pyplot.scatter(height,weight)
pyplot.show()
```

```python
height = [20,30,40,50,60]
weight = [39,29,10,11,12]

pyplot.scatter(height,weight,color = 'red')
pyplot.xlabel("height in cm")
pyplot.ylabel("weight in kg")
pyplot.title("height and weight")
pyplot.show()
```

```python
velocity = [10,20,30,40,50]

displacemet = [25,35,45,55,65]

pyplot.scatter(velocity,displacemet,color = 'red')
pyplot.xlabel("velocity in m/s")
pyplot.ylabel("displacemet in m")
pyplot.title("velocity and displacement")
pyplot.show()

```

```python
velocity = [10,20,30,40,50]

displacemet = [25,35,45,55,65]

pyplot.plot(velocity,displacemet,color = 'yellow')
pyplot.xlabel("velocity in m/s")
pyplot.ylabel("displacemet in m")
pyplot.title("velocity and displacement")
pyplot.show()
```

```python
x1 = [2,4,6,8,10]

y1 = [3,6,9,12,15]

x2 = [1,4,8,12]

y2 = [3,5,7,9]

pyplot.plot(x1,y1,color ='red')
pyplot.plot(x2,y2,color = 'green')
pyplot.xlabel("x axis")
pyplot.ylabel('y axis')
pyplot.title("x1 & y1  and x2 and y2")
pyplot.show()
```

```python
# BAR CHART

x = [5,7,8,9,10]

y = [3,7,4,8,3]

pyplot.bar(x,y)

pyplot.xlabel("x axis")
pyplot.ylabel('y axis')
pyplot.title("x and y ")

pyplot.show()
```

```python
subjects = ["maths","physics","chemistry","biology"]

marks = [70,85,93,90]

pyplot.bar(subjects,marks,color ='yellow')
pyplot.xlabel("subjects",fontsize = 15,color = 'blue')
pyplot.ylabel('marks',fontsize = 15,color = 'blue')
pyplot.title("subject and mark")
pyplot.show()
```

```python
# histogram 


marks = [0,1,6,7,8,12,15,20,25,21,22,22,35,36,38,49,48,50]

pyplot.xlabel('marks')

pyplot.ylabel('count')

pyplot.title('student marks')

pyplot.hist(marks)

pyplot.show()
```

```python
x = [1,1,2,2,2,4,4,4,5,5,5,5,6,6,6,6]

pyplot.hist(x)
pyplot.xlabel('x axis')
pyplot.ylabel('count')
pyplot.title("histogram of x")
pyplot.show()
```

```python
# piechart 
# ========

activities = ['sleeping','travel','work','others']

hours = [8,4,8,4]

pyplot.pie(hours,labels= activities)

pyplot.show()
```

```python
activities = ['sleeping','travel','work','others']

hours = [8,4,8,4]

colors = ['red','yellow','green','violet']

explode = (0.1,0,0,0)

pyplot.pie(hours,labels= activities,colors= colors,explode=explode,autopct='%1.1f%%',startangle= 90,shadow=True)

pyplot.show()
```
