```python
import pandas
import numpy
```

```python
data = pandas.read_csv("sales_data_clean.csv")
df = pandas.DataFrame(data)
df
```

```python
from matplotlib import pyplot

x = df['month_number']
y = df['total_profit']

pyplot.plot(x,y,color = 'red')
pyplot.xlabel('month_number')
pyplot.ylabel('total_profit')
pyplot.title("month_number / total_profit")
pyplot.show()
```

```python
product_sales = [df['facecream'].sum(),
                 df['facewash'].sum(),
                 df['bathsoap'].sum(),
                 df['moisturizer'].sum(),
                 df['shampoo'].sum(),
                 df['toothpaste'].sum(),]

items = ['facecream','facewash','toothpaste','bathsoap','shampoo','moisturizer']

colors = ['red','yellow','green','violet','brown','blue']

explode = (0.1,0,0,0,0,0)

pyplot.pie(product_sales,labels=items,colors=colors,explode=explode,autopct='%1.1f%%',startangle= 90, shadow= True)

pyplot.show()
```
