KNN (using distance formula)   

KNN Algorithm steps
===================

step1 : Select the number of k neighbours

step2 : Calculate the euclidean distance of the new value (x1,y1) from all the available datapoints

step3 : Take the k (given value in step 1) nearest neighbours as per the calculated distance

step4 : Among these k neighbours selected count the number of datapoints in each category (count of pass and fail)

step5 : Assign the new data point to the category for which has maximum count

step6 : model has been created


Euclidean distance   >>>>  sqrt ( (xi - x_new)^2 + (yi - y_new)^2) 


(xi , yi)   coordinate of points in the graph (existing students mark and attendance)
(x_new , y_new) coordinate of the new point which need to be predicted
