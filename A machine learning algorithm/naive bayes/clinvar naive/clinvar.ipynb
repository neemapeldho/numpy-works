```python
import pandas
import numpy
```

```python
data = pandas.read_csv("clinvar_conflicting.csv")
df = pandas.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df = df.drop(columns=["CLNDISDBINCL","CLNDNINCL","CLNSIGINCL","CLNVI","SSR","INTRON","DISTANCE","BAM_EDIT","SIFT","MOTIF_NAME","MOTIF_POS","HIGH_INF_POS","MOTIF_SCORE_CHANGE","BLOSUM62"])
df
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=["PolyPhen"],axis=0)
df.isna().sum()
```

```python
df.dtypes
```

```python
df["LoFtool"]=df["LoFtool"].fillna(df["LoFtool"].mean())
df["CADD_PHRED"]=df["CADD_PHRED"].fillna(df["CADD_PHRED"].mean())
df["CADD_RAW"]=df["CADD_RAW"].fillna(df["CADD_RAW"].mean())
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in df.columns:
    if df[i].dtype == "object":
        df[i] = le.fit_transform(df[i].astype(str)) 
        # .astype(str) ensures all entries are treated as strings, so no type conflict occurs.
    else:
        pass
```

```python
df.columns
```

```python
Y = df.iloc[:,28]
Y
```

```python
df_copy = df
df = df.drop(columns=["PolyPhen"],axis=1)
```

```python
X = df
X
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb = nb.fit(x_train,y_train)

y_pred = nb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_pred))
```
