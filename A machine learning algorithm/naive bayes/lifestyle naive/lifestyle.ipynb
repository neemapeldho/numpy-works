```python
import pandas
import numpy
```

```python
data = pandas.read_csv("Lifestyle.csv")
df = pandas.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=["married"],axis=1)
df
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["exercise"] = le.fit_transform(df["exercise"])
df["sugar_intake"] = le.fit_transform(df["sugar_intake"])
df["smoking"] = le.fit_transform(df["smoking"])
df["alcohol"] = le.fit_transform(df["alcohol"])
df["profession"] = le.fit_transform(df["profession"])
df["health_risk"] = le.fit_transform(df["health_risk"])
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb = nb.fit(x_train,y_train)

y_pred = nb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_pred))
```
