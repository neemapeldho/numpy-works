```python
import pandas
import numpy
```

```python
data = pandas.read_csv("medical_insurance.csv")
df = pandas.DataFrame(data)
df
```

```python
df.dtypes
```

```python
df = df.drop(columns=["person_id","education","region","urban_rural","marital_status","household_size"],axis=1)
df
```

```python
df.isna().sum()
```

```python
df["alcohol_freq"] = df["alcohol_freq"].fillna(df["alcohol_freq"].mode()[0])
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df["sex"] = le.fit_transform(df["sex"])
df["smoker"] = le.fit_transform(df["smoker"])
df["employment_status"] = le.fit_transform(df["employment_status"])
df["alcohol_freq"] = le.fit_transform(df["alcohol_freq"])
df["plan_type"] = le.fit_transform(df["plan_type"])
df["network_tier"] = le.fit_transform(df["network_tier"])
```

```python
df.dtypes
```

```python
Y = df.iloc[:,46]
df = df.drop(columns=["is_high_risk"],axis=1)
X = df
Y
X
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn = knn.fit(x_train,y_train)

y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_pred))
```
