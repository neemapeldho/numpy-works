```python
import pandas
import numpy
```

```python
data = pandas.read_csv('synthetic_dataset.csv')
df = pandas.DataFrame(data)
df
```

```python
# Display all products with a Price greater than 500.

df[df["Price"]>500]
```

```python
# Select products with a Rating of 4.5 or higher.

df[df["Rating"] >= 4.5]
```

```python
# Find all items that have a Discount greater than 20% but Price less than 1000.

df[(df["Discount"] > 20) & (df['Price'] < 1000)]
```

```python
# Display only Price and Rating columns for the first 10 items.

df.loc[:,['Price','Rating']].head(10)
```

```python
# Find the product(s) with the maximum and minimum Price.

df[df['Price'] == (df['Price']).min()]
```

```python
df[df['Price'] == (df['Price']).max()]
```
