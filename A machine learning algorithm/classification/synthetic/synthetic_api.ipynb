```python
import pandas
import numpy
```

```python
data = pandas.read_csv('synthetic_dataset.csv')
df = pandas.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
df = df.drop('Category',axis = 1)
df
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset= 'Rating')
df
```

```python
df.isna().sum()
```

```python
df['Price'] = df['Price'].fillna(df['Price'].mean())
df.isna().sum()
```

```python
df['Discount'] = df['Discount'].fillna(df['Discount'].mean())
df.isna().sum()
```

```python
df['Stock'] = df['Stock'].fillna(df['Stock'].mode()[0])
df.isna().sum()
```

```python
df
```

```python
df['Stock'] = df['Stock'].map({'In Stock' : 0,'Out of Stock' : 1})   # encoding  >>>> data preprocessing
df
```
