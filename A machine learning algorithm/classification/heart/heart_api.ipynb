while filling the missing values if more than 50 % row missing skip the column 

```python
# fill the missing values
df['trestbps'].fillna(df['trestbps'].mean(),inplace=True)
df.isna().sum()
```

```python
df['restecg'].fillna(df['restecg'].mean(),inplace=True)
df.isna().sum()
```

```python
df['thalach'].fillna(df['thalach'].mean(),inplace=True)
df.isna().sum()
```

```python
df['ca'].fillna(df['ca'].mean(),inplace=True)
df.isna().sum()
```

```python
df['thal'].fillna(df['thal'].mean(),inplace=True)
df.isna().sum()
```

```python
# diaplay all male patients(sex = 1) older than 50 years
# df['sex']
# df[df['sex'] == 1]
df[(df['sex'] == 1) & (df['age'] > 50)] 
```

```python
# find patients who have chest pain type (cp) = 0 and target = 1
df[(df['cp'] == 0) & (df['target'] == 1)]
```

```python
# count how many patients have fasting blood sugar (fbs) = 1

(df['fbs'] == 1).sum()
```

```python
# find the average cholestrol by gender
# df.groupby('sex').mean()
df.groupby('sex')['chol'].mean()
```

```python
# find the average thalach by gender

df.groupby('sex')['thalach'].mean()
```

```python
# count how many patients belong to each chest pain type (cp) . cp = 0 count cp = 1 count

print((df['cp'] == 1).sum())
print((df['cp'] == 0).sum())
```

```python
# find the top 5 patients with the highest cholestrol values
# highest values in column = outliers / extrem
# .nlargest() → a pandas function that returns the top n rows with the largest values in one or more specified columns

df.nlargest(n = 5,columns = ['chol'])
```

```python
# create a new column bp_category with values "normal" if trestbps < 120 else "high"
# .apply(...) → Runs a function on each value of that column.

df['bp_category'] = df['trestbps'].apply(lambda a :"normal" if a < 120 else "high")
df
```
