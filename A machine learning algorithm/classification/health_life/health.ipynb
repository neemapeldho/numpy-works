```python
# find the maximum and minimum dailysteps

print(df['daily_steps'].max())
print(df['daily_steps'].min())
```

```python
# select age ,gender ,bmi of people aged more than 40

df.loc[df['age'] > 40,['age','gender','bmi']] # used to select some columns and get in a new array

# loc - Access a group of rows and columns by label(s) or a boolean array.  loc[row,column]
```

```python
# select gender,daily_steps,sleep_hours where sleep_hours less than 6

df.loc[df['sleep_hours'] < 6,['gender','daily_steps','sleep_hours']]
```

```python
# display age,bmi,calories_consumed where bmi > 30

df.loc[df['bmi'] > 30,['age','bmi','calories_consumed']]
```

```python
# dispaaly age,daily_steps,calories_consumed foe smokers (smoker = 1)

df.loc[df['smoker'] == 1,['age','daily_steps','calories_consumed']]
```

```python
# display age,daily_steps,bmi of people having bmi above 28 and disease = 1

df.loc[(df['bmi'] > 28) & (df['disease_risk'] == 1) ,['age','daily_steps','bmi']]
```

```python
# diaplay age,daily steps,bmi of people having bmi above 28 or smoker = 1

df.loc[(df['bmi'] > 28) | (df['smoker'] == 1),['age','daily_steps','bmi']]
```

```python
# display daily_steps,cholesterol,bmi of people having smoker = 1 and alcohol= 1

df.loc[(df['smoker'] == 1) & (df['alcohol'] == 1),['daily_steps','cholesterol','bmi']]
```

```python
# display age,daily_steps,cholesterol of people having resting_hr above 90 or systolic_bp > 150

df.loc[(df['resting_hr'] > 90) | (df['systolic_bp'] > 150) ,['age','daily_steps','cholesterol']]
```
