```python
import pandas
import numpy
```

```python
data = pandas.read_csv('loan_train.csv')
df = pandas.DataFrame(data)
df
```

```python
# target >>> status
# input features >>> Dependents	Self_Employed	Applicant_Income	Coapplicant_Income	Loan_Amount	Term	Credit_History	Area
```

```python
df = df.drop(columns=['Gender','Education'],axis=1)  
df
# axis=1  Operate along columns
```

```python
df.isna().sum()
```

```python
df['Dependents'] =df['Dependents'].fillna(df['Dependents'].mode()[0])
df.isna().sum()
```

```python
df['Married'] =df['Married'].fillna(df['Married'].mode()[0])
df.isna().sum()
```

```python
df['Self_Employed'] =df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df.isna().sum()
```

```python
df['Term'] =df['Term'].fillna(df['Term'].mean())
df.isna().sum()
```

```python
df['Credit_History'] =df['Credit_History'].fillna(df['Credit_History'].mean())
df.isna().sum()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Married'] = le.fit_transform(df['Married'])
df['Dependents'] = le.fit_transform(df['Dependents'])
df['Self_Employed'] = le.fit_transform(df['Self_Employed'])
df['Area'] = le.fit_transform(df['Area'])
df['Status'] = le.fit_transform(df['Status'])
df
```

```python
df.dtypes
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
X
Y 
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size= 0.3)

X_train
X_test
Y_train
Y_test
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() 
scaler.fit(X_train) 

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier   

knn = KNeighborsClassifier()   

knn = knn.fit(X_train,Y_train)     

y_pred = knn.predict(X_test)     

y_pred
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_pred,Y_test))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,y_pred)

cm 
```
