```python
import pandas
import numpy
```

```python
data = pandas.read_csv("heart1.csv")
df = pandas.DataFrame(data)
df
```

```python
# target >>> HeartDisease
# input features >>> Age	Sex	ChestPainType	RestingBP	Cholesterol	FastingBS	RestingECG	MaxHR	ExerciseAngina	Oldpeak	ST_Slope
```

```python
df = df.drop(columns=['Age','Sex'],axis=1)    # df = ... or inplace = True for change in data frame
df
```

```python
df.isna().sum()    # no missing values
```

```python
# check any object present and if present convert to numerical - encoding
# encoding >>>> convert to appropriate integer value

from sklearn.preprocessing import LabelEncoder      # ML concept in sklearn

le = LabelEncoder()

df['ChestPainType'] = le.fit_transform(df['ChestPainType'])   # fit >> collect   , transform >> transform  or fit_transform
df['RestingECG'] = le.fit_transform(df['RestingECG'])
df['ExerciseAngina'] = le.fit_transform(df['ExerciseAngina'])
df['ST_Slope'] = le.fit_transform(df['ST_Slope'])
df

```

```python
X = df.iloc[:,:-1] # input features
Y =df.iloc[:,-1]   # ouput / result
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size= 0.3)
```

```python
X_train
```

```python
X_test
```

```python
# only x need to scale , y no need to scale having only 1 column

from sklearn.preprocessing import StandardScaler # Z = (x-mean)/std

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)    # model need to be predict the result for the unseen data - Data leakage

```

```python
X_train
```

```python
# Data preprocessing steps:
# ========================

# missing values handle
# outliers
# data transform
# scaling
# data reduction
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

model = knn.fit(X_train,Y_train)

y_pred = model.predict(X_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_pred,Y_test))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,y_pred)

cm
```
