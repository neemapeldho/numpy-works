```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("hypothyroid.csv")
df = pd.DataFrame(data)
df
```

```python
df = df.replace("?",np.nan)
df
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=["age","sex","TBG"],axis=1)
df
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df["TSH"] = df["TSH"].fillna(df["TSH"].mode()[0])
df["T3"] = df["T3"].fillna(df["T3"].mode()[0])
df["TT4"] = df["TT4"].fillna(df["TT4"].mode()[0])
df["T4U"] = df["T4U"].fillna(df["T4U"].mode()[0])
df["FTI"] = df["FTI"].fillna(df["FTI"].mode()[0])
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in list(df):

    if df[i].dtype == "object":

        df[i] = le.fit_transform(df[i].astype(str))
```

```python
df.dtypes
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

obj = DecisionTreeClassifier(criterion="entropy")  , # select entropy or default take gini
obj = obj.fit(x_train,y_train)

y_pred = obj.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_pred))
```
