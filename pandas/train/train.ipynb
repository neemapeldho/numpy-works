```python
import pandas
import numpy
```

```python
data = pandas.read_csv("train.csv")
df = pandas.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
# convert order date and ship date to datetime

df["Ship Date"] = pandas.to_datetime(df["Ship Date"],format="%d/%m/%Y")
df["Order Date"] = pandas.to_datetime(df["Order Date"],format="%d/%m/%Y")
```

```python
df.dtypes # Show the data types of all columns
```

```python
# Load the CSV file and display the first 10 rows.

df.head(10)
```

```python
#Display the total number of rows and columns.

df.shape
```

```python
# Remove all duplicate rows and show the new shape.

df = df.drop_duplicates()
df.shape
```

```python
# Rename these columns for readability:
    # Row ID → row_id
    # Order ID → order_id

df.rename(columns={"Row ID" : "row_id"},inplace=True)
df.rename(columns={"Order ID" : "order_id"},inplace=True)
df.dtypes
```

```python
# Sort the dataset by Sales (descending).
 
df.sort_values(by="Sales",ascending=False)
```

```python
# Filter all orders where Sales > 500.

df.loc[df["Sales"] > 500]
```

```python
# Get all orders shipped using “Second Class” mode.

df.loc[df["Ship Mode"] == "Second Class"]
```

```python
# Select all orders from the State = “California”.

df[df["State"] == "California"]
```

```python
# Select all orders from Consumer Segment with sales above 200.

df[(df["Segment"] == "Consumer") &  (df["Sales"] > 200)]
```

```python
# Display all orders where the category is “Furniture” and Region = “West”.

df[(df["Category"] == "Furniture") & (df["Region"] == "West")] 
```

```python
# Find all orders where the ship date is after the order date.

df[df["Ship Date"] > df["Order Date"]]
```

```python
# Create a new column Order_Year from Order Date.

df["Order_Year"] = df["Order Date"].apply(lambda y:y.year)
df
```

```python
# Create a new column Shipping_Time_days = Ship Date – Order Date.

df["Shipping_Time_days"] = df["Ship Date"] - df["Order Date"]
df["Shipping_Time_days"].apply
df
```

```python
# Create a new column Profit_Category:
    # Sales > 500 → “High”
    # Sales between 100–500 → “Medium”
    # Sales < 100 → “Low”

df["Profit_Category"] = df["Sales"].apply(lambda a: "High" if a > 500 else "Medium" if a > 100 else "Low")
df
```

```python
# Create a new column that extracts first name of the customer.

df["first name"] = df["Customer Name"].apply(lambda a:a.split(" ")[0])
df
```

```python
# Create a new column discount_amount (assume discount = 10% of sales).
df["discount_amount"] = df["Sales"].apply(lambda a: (10/100) * a)
df
```

```python
# Convert all values in City column to lowercase.

df["City"] = df["City"].str.lower()
df
```

```python
# Replace spaces in Product Name with underscores.

# df.rename(columns={"Product Name":"Product_Name"},inplace=True)

df["Product Name"] = df["Product Name"].str.replace(" ","_")
df
```
